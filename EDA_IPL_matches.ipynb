```python
#Importing Libraries
import pandas as pd
import numpy as np
```

```python
#Importing file and reading it
df = pd.read_csv("Desktop/matches.csv.csv")
df.head()
```

```python
#Check which columns has more missing value or NaN value and drop it.
df.info() #Since, in the below output umpire3 has very less value as compared to other col., we can also check the uniquness
          #by df['umpire3'].unique and drop it.
```

```python
#Dropping the umpire3 column
df.drop(['umpire3'],axis=1)
```

```python
#Total number of matches played at each season
df['season'].value_counts()

#                           OR
#If you want to represent it in more better way, you can write below a bit complex code
#for i in range(df['season'].nunique()):  --> nunique will give total number of season played till date
#    print("Total number of matches in", df['season'].unique()[i], " : ", df[df['season']==df['season'].unique()[i]].shape[0])
```

```python
#No. of times a player got player_of_the_match award
df['player_of_match'].value_counts()
```

```python
#Maximum number of matches won by which team?
df['winner'].value_counts().idxmax()
```

```python
#Find team which has won matches by huge margin, a win by more than 135 run or more than 8 wicket will be considered as big win as per our condition
big_win = df[(df['win_by_runs'] >= 135) | (df['win_by_wickets']>=9)]
big_win
```

```python
#Maximum and minimum win_by_wicket at each city
for i in range(df['city'].nunique()):
    print("Win by most number of wicket at ", df['city'].unique()[i], df.win_by_wickets[df['city']==df['city'].unique()[i]].max())
    print("Win by least number of wicket ", df['city'].unique()[i], df.win_by_wickets[df['city']==df['city'].unique()[i]].min())
```

```python
#Maximum margin of win_by_run at each city
for i in range(df['city'].nunique()):
    print("Maximux margin of run win at ", df['city'].unique()[i], df.win_by_runs[df['city']==df['city'].unique()[i]].max())
```
